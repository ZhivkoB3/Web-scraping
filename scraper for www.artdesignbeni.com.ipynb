```python
import requests
from requests import get
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
from pandas import DataFrame
import numpy as np
import csv
from time import sleep
from random import randint
import re
```

```python
url_to_scrape = "https://www.artdesignbeni.com/bizuta-ot-smola-1.html"
page_html = requests.get(url_to_scrape)
html_soup = BeautifulSoup(page_html.content, 'html.parser')

product_data = []
for div in html_soup.find_all('div', class_='gs-item-text'):
    product = div.find('a').contents[0]
    price = div.find('span', class_='gs-new-price').text
    scrapped_data = [product, price]
    product_data.append(scrapped_data)
```

```python
header = ['Модел', 'Цена']
with open('artdesign.csv', 'w', encoding = 'utf-8', newline = '') as file:
    writer = csv.writer(file, quoting = csv.QUOTE_ALL)
    writer.writerow(i for i in header)
    for row in product_data:
        writer.writerow(row)
```

```python
df = pd.read_csv('artdesign.csv', decimal = ',')
```

```python
df = df.apply(lambda S:S.str.strip('\n\t\t\t\t\t\t\t\t\t\t\t'))
df = df.apply(lambda S:S.str.strip('лв.'))
df['Цена'] = df['Цена'].str.replace(",",".").astype(float)
```
